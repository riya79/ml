## Step 4 : Train-Test Split and feature scaling

```python
data_x = df[['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
data_y = df['Class']
```

```python
data_x.shape
```

```python
data_y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(data_x,data_y,test_size=0.2,
                                                         random_state=42)
```

## Step 6 : Model Building

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

reg = DecisionTreeRegressor(max_depth=6)
reg.fit(X_train, y_train)
```

```python
# Evaluate the models
print("Train Accuracy :", reg.score(X_train, y_train))
print("Test Accuracy :", reg.score(X_test, y_test))
```
