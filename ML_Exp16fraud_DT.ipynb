# Step 3 : Visualizing the data

```python
# checking for correlation
print(df.corr())
plt.figure(figsize=(30, 10))
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
x = df[['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
y = df['Class']
```

```python
x.info()
```

```python
y.unique()
```

# Step 5 : Train-Test Split and feature scaling

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

# Step 6 : Model Building 

## Decision Tree regressor

```python
from sklearn.tree import DecisionTreeRegressor

reg = DecisionTreeRegressor(max_depth=6)
```

```python
reg.fit(x_train, y_train)
```

```python
print("Train Accuracy :", reg.score(x_train, y_train))
print("Test Accuracy :", reg.score(x_test, y_test))
```
